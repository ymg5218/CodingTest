```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = [1, 2, 3, 4 , 5]
plt.plot(y)
plt.show()
```

```python
y = [13, 16, 15, 18, 16, 17, 16]
plt.plot(y);
```

```python
x = ['a', 'b', 'c', 'd', 'e', 'f', 'g']
plt.plot(x, y);
plt.xlabel('name')
plt.ylabel('score')
```

```python
y2 = [20, 10, 12, 9, 23, 19, 15]
plt.plot(x, y, label='math');
plt.plot(x, y2, label='science');
plt.xlabel('name')
plt.ylabel('score');
plt.legend(loc="upper left")
plt.title('Student State');
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.axis([0, 6, 0, 30])
plt.plot(x, y, 'bo');
```

```python
points = np.array([[1, 1], [1, 2], [1, 3], [2, 1], [2, 2], [2, 3], [3, 1], [3, 2], [3, 3]])
p = np.array([2.5, 2])
plt.plot(points[:, 0], points[:, 1], "ro")
plt.plot(p[0], p[1], "bo")
```

```python
x = np.linspace(0, 5, 20)
y = x**2
plt.plot(x, y, "bs-", linewidth=2, markersize=5);
```

```python
x = np.linspace(0, 10, 20)
y1 = x**2.0
y2 = x**1.5
plt.plot(x, y1, 'bo-', linewidth=2, markersize=5, label="First")
plt.plot(x, y2, 'gs-', linewidth=2, markersize=5, label="Second")
plt.xlabel("X")
plt.ylabel("Y")
plt.axis([-1, 10.5, -5, 105])
plt.legend(loc="upper left")
plt.savefig("mplot.pdf");
```

```python
x = np.logspace(-1, 1, 20)
print(x)
y1 = x**2.0
y2 = x**1.5
plt.plot(x, y1, 'bo-', linewidth=2, markersize=5, label="First")
plt.plot(x, y2, 'gs-', linewidth=2, markersize=5, label="Second")
plt.xlabel("X")
plt.ylabel("Y")
plt.axis([-1, 10.5, -5, 105])
plt.legend(loc="upper left")
plt.savefig("mplot2.pdf");
```

```python
plt.style.use('default')
plt.rcParams['font.size'] = 12
x = np.random.standard_normal(1000)
y = np.random.standard_normal(10000)
z = np.random.standard_normal(100000)
plt.hist(x, bins=100, density=True, alpha=0.5,histtype='step', label='size=1000')
plt.hist(y, bins=100, density=True, alpha=0.75, histtype='step', label='size=10000')
plt.hist(z, bins=100, density=True, alpha=1, histtype='step', label='size=100000')
plt.legend()
plt.show()
```
